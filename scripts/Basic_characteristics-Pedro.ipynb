```python
from __future__ import division
import os
import ast
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
#import distance
from collections import OrderedDict
import networkx as nx
import seaborn as sns
sns.set(rc={'axes.facecolor':'white'})

%matplotlib inline
```

### 1. Loading in data

```python
#Hypothetical chronological order
aas = ['gly','ala','asp','val','pro','ser','glu','leu','thr','arg','gln','ile','asn','his','lys','cys','phe','tyr','met','trp']
```

```python
trnas = OrderedDict()

for aa in aas:
    trnas[aa] = pd.DataFrame.from_csv('../data/' + aa + '.txt', sep = '\t')
```

### 2. Building graphs

```python
H = {}

for aa in trnas:
    H[aa] = nx.Graph()
    H[aa].size = {}
    H[aa].color = {}
    H[aa].width = {}

    for i in trnas[aa].index:

        H[aa].add_node(i)

    for i in trnas[aa].index:
        for j in ast.literal_eval(trnas[aa]['Adjacency'].ix[i]):
            H[aa].add_edge(trnas[aa].ix[i].name, trnas[aa].ix[j].name)
```

### 3. Degree

#### 3.1. Average degrees

```python
avgDegree = []

for aa in aas:
    avgDegree.append(np.mean(nx.degree(H[aa]).values()))
    
plt.figure(figsize=[12,7])
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
sns.stripplot(aas, avgDegree,size=10)
for i in range(20):
    plt.plot([i]*1000, np.linspace(-0.1,1.5,1000), '--', color='grey', lw=0.5)
plt.title('Average degree');
```

tARN's gly, ala, and glu are the only networks with more links than nodes. Most of them have a number of links between 60% the number of nodes. Furthermore, even the nodes with more links such as gly, ala or glu, they all have around 50% of the nodes with any connections, as we show below in the degree distribution.

#### 3.2. Degree distribution

```python
plt.figure(figsize=(4*5+3,5*5))
count=1
    
for aa in trnas:
    plt.subplot(5,4,count)
    degreeHist = nx.degree_histogram(H[aa])
    plt.bar([x for x in range(len(degreeHist))], degreeHist, color='k', log=True)
    plt.title(aa)
    plt.xticks([x for x in range(20)])
    for i, v in enumerate(degreeHist):
        if v:
            plt.annotate(v, xy=(i, v))
    plt.xlim(-1,20)
    count+=1
```

As we commented before, with this plot we can see that most of the nodes in each of our complex networks have between 0 and 4 links.

### 4. Connected components

```python
connectedComponents=[]
for aa in aas:
    counter = 0
    for i in nx.connected_components(H[aa]):
        #if len(i)>1:
        counter += 1
    connectedComponents.append(counter)
```

Example of number of nodes per component histogram.

```python
tRNA = 'his'
histogram_of_components = []
for comp in nx.connected_components(H[tRNA]):
    histogram_of_components.append( len(comp) )

m = min(histogram_of_components)
M = max(histogram_of_components)
print m,M


def hist(a, aa):
    n = np.bincount(a)
    
    plt.bar([x for x in range(len(n))], n, color='k', log=True)
    plt.title(aa)
    #plt.xticks([x for x in range(20)])
    ticks = []
    for i, v in enumerate(n):
        if v:
            plt.annotate(v, xy=(i, v))
            ticks.append(i)
    plt.xticks(ticks)
    plt.xlim(0,max(ticks))
    plt.ylabel('Counts')
    plt.xlabel('# nodes per component')

hist(histogram_of_components, tRNA)
```

```python
plt.figure(figsize=(4*5+3,5*5))
count=1
  
for aa in trnas:
    plt.subplot(5,4,count)
    
    histogram_of_components = []
    for comp in nx.connected_components(H[aa]):
        histogram_of_components.append( len(comp) )
    
    hist(histogram_of_components, aa)
    count+=1
```

In some networks there exist a component with a very large number of nodes, such as in [gly, ala, val, glu], where the first three contains a component with more than 470 nodes and the last one contains a component with 173 nodes.

```python
a = [len(trnas[x]) for x in trnas]

plt.plot(connectedComponents, a, "o")

for i, v in enumerate(trnas):
    plt.annotate(v, xy=(connectedComponents[i], len(trnas[v])+100))

plt.ylabel('# nodes')
plt.xlabel('# of connected components')

# fit a line for this points
slope, intercept  = np.polyfit(connectedComponents, a, 1)
abline_values = [slope * i + intercept for i in connectedComponents]
plt.plot(connectedComponents, abline_values, 'b')


#plt.plot(connectedComponents,[len(x) for x in trnas])
```

Thanks to the above plot we can claim that there is a linear correlation between the number of sequences of a tRNA (nodes) and the number of connected components. It must be due to the sparsity of the links between nodes: As much number of nodes, much number of nodes with any link.

```python
plt.figure(figsize=[12,7])
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
sns.stripplot(aas, connectedComponents, size=10)
plt.title('Connected components');
```

```python
connectedComponents_sorted = np.asarray(connectedComponents).copy()
sort_idx = connectedComponents_sorted.argsort()[::-1]

plt.figure(figsize=[12,7])
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
sns.stripplot(np.asarray(aas)[sort_idx], connectedComponents_sorted[sort_idx], size=10)
for i in range(20):
    plt.plot([i]*1000, np.linspace(-10,420,1000), '--', color='grey', lw=0.5)
plt.title('Connected components');
```

A large number of components means that the links are more sparse. In the other side, this tARN's which have a higher number of components are also the older ones, so, hipotetically they had more time to mutate and hence to connect more nodes. This is esxactly the opposite od what we see above, unless almost all mutations have been new nodes instead of existing ones.

Let us see how this distribution changes if we normalize by the number of nodes, as we know from the discussion above, that for larger number of nodes, larger number of components, and we are more interested in the percentatge.

```python
connectedComponents_norm = np.asarray(connectedComponents).copy().astype(np.float)

for i in range(len(aas)):
    connectedComponents_norm[i] = connectedComponents_norm[i]/len(trnas[aas[i]])
```

```python
plt.figure(figsize=[12,7])
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
sns.stripplot(aas, connectedComponents_norm, size=10)
plt.title('Connected components normalized');
```

### If connected components impose the antiquity...

```python
connectedComponents_norm_sorted = np.asarray(connectedComponents_norm).copy()
sort_idx = connectedComponents_norm_sorted.argsort()[::-1]

plt.figure(figsize=[12,7])
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
sns.stripplot(np.asarray(aas)[sort_idx], connectedComponents_norm_sorted[sort_idx], size=10)
plt.title('Connected components');
```

# Minimum Distance required to have a completely connected graph

```python
def graph_connected(G, verbose=False):
    """ True if the graph G is connected """
    all_connected = False
    connectedComponents=[]
    
    counter = 0
    for i in nx.connected_components(G):
        counter += 1
    if counter == 1:
        all_connected = True
    
    if verbose:
        print counter
    return all_connected

tRNA = 'his'
print ('test for a non-connected graph')
print graph_connected(H[tRNA])

print ('\ntest for a connected graph')
tmp = nx.Graph()
tmp.add_edge(1,2)
tmp.add_edge(3,4)
tmp.add_edge(2,4)

print graph_connected(tmp, True)
```

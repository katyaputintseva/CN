```python
from __future__ import division
import distance
import os
import ast
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import distance
from collections import OrderedDict
import networkx as nx
import seaborn as sns
sns.set(rc={'axes.facecolor':'white'})

%matplotlib inline
```

## 1. Loading in data

```python
#Hypothetical chronological order
aas = ['gly','ala','asp','val','pro','ser','glu','leu','thr','arg','gln','ile','asn','his','lys','cys','phe','tyr','met','trp']
```

```python
trnas = OrderedDict()

for aa in aas:
    trnas[aa] = pd.DataFrame.from_csv('../data/' + aa + '.txt', sep = '\t')
```

***

## 2. Different types of tRNAs studied separately

### 2.1. Building graphs

```python
H = {}

for aa in trnas:
    H[aa] = nx.Graph()

    for i in trnas[aa].index:

        H[aa].add_node(i)

    for i in trnas[aa].index:
        for j in ast.literal_eval(trnas[aa]['Adjacency'].ix[i]):
            H[aa].add_edge(trnas[aa].ix[i].name, trnas[aa].ix[j].name)
```

### 2.2. Degree

#### 2.2.1. Average degrees

```python
avgDegree = []

for aa in aas:
    avgDegree.append(np.mean(nx.degree(H[aa]).values()))
    
plt.figure(figsize=[12,7])
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
sns.stripplot(aas, avgDegree,size=10)
for i in range(20):
    plt.plot([i]*1000, np.linspace(-0.1,1.5,1000), '--', color='grey', lw=0.5)
plt.title('Average degree');
```

#### 2.2.2. Degree distribution

```python
plt.figure(figsize=(4*5+3,5*5+7))
count=1
    
for aa in trnas:
    plt.subplot(5,4,count)
    degreeHist = nx.degree_histogram(H[aa])
    degreeHist = [x/sum(degreeHist) for x in degreeHist]
    plt.scatter([x for x in range(len(degreeHist))], degreeHist, color='k')
    plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
    plt.title(aa, fontsize=20)
    plt.xlabel('k')
    plt.ylabel('p(k)')
    plt.ylim(-0.02,1)
    count+=1
```

### 2.3. Connected components

```python
connectedComponents=[]
for aa in aas:
    counter = 0
    for i in nx.connected_components(H[aa]):
        if len(i)>1:
            counter += 1
    connectedComponents.append(counter)
```

```python
plt.figure(figsize=[12,7])
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
sns.stripplot(aas, connectedComponents, size=10)
for i in range(20):
    plt.plot([i]*1000, np.linspace(-10,420,1000), '--', color='grey', lw=0.5)
plt.title('Connected components');
```

### 2.5. Average neighbor degree

```python
plt.figure(figsize=(4*5+3,5*5+7))
count=1

for aa in aas:
    
    avgNeighborDegree = nx.average_neighbor_degree(H[aa])
    avgNeighborDegree = pd.DataFrame.from_dict(avgNeighborDegree, orient='index')
    avgNeighborDegree.columns = ['avgNeighborDegree']
    nodeDegree = nx.degree(H[aa])
    nodeDegree = pd.DataFrame.from_dict(nodeDegree, orient='index')
    nodeDegree.columns = ['nodeDegree']
    joinedDF = nodeDegree.join(avgNeighborDegree)
    plt.subplot(5,4,count)
    plt.plot(joinedDF.avgNeighborDegree, joinedDF.nodeDegree,  'ok', alpha = 0.6)
    plt.title(aa, fontsize=20)
    plt.ylabel('k')
    plt.xlabel('Average neighbor degree')
    plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')

    count+=1;
```

### 2.6. Node degree vs clustering coefficient

```python
plt.figure(figsize=(4*5+3,5*5+5))
count=1

for aa in aas:
    
    clusteringCoeff = nx.clustering(H[aa])
    clusteringCoeff = pd.DataFrame.from_dict(clusteringCoeff, orient='index')
    clusteringCoeff.columns = ['clusteringCoeff']
    nodeDegree = nx.degree(H[aa])
    nodeDegree = pd.DataFrame.from_dict(nodeDegree, orient='index')
    nodeDegree.columns = ['nodeDegree']
    joinedDF = nodeDegree.join(clusteringCoeff)
    plt.subplot(5,4,count)
    plt.plot(joinedDF.nodeDegree, joinedDF.clusteringCoeff, 'ok', alpha = 0.6)
    plt.title(aa)
    plt.ylabel('Clustering coefficient')
    plt.xlabel('k')
    plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')

    count+=1;
```

## 3. Studying a meta graph (all types of tRNAs combined together)

### 3.1. Building graph

```python
for aa in trnas:
    G = nx.Graph()

    for i in trnas[aa].index:

        G.add_node(i)

    for i in trnas[aa].index:
        for j in ast.literal_eval(trnas[aa]['Adjacency'].ix[i]):
            G.add_edge(trnas[aa].ix[i].name, trnas[aa].ix[j].name)
```

### 3.2. Degree distribution

```python
plt.figure(figsize=[10,8])

degreeHist = nx.degree_histogram(G)
degreeHist = [x/sum(degreeHist) for x in degreeHist]
plt.scatter([x for x in range(len(degreeHist))], degreeHist, color='k')
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
plt.title('Meta Graph', fontsize=20)
plt.xlabel('k')
plt.ylabel('p(k)')
plt.ylim(-0.02,1)
```

### 3.3. Node degree vs. average neighbor degree

```python
avgNeighborDegree = nx.average_neighbor_degree(G)
avgNeighborDegree = pd.DataFrame.from_dict(avgNeighborDegree, orient='index')
avgNeighborDegree.columns = ['avgNeighborDegree']
nodeDegree = nx.degree(G)
nodeDegree = pd.DataFrame.from_dict(nodeDegree, orient='index')
nodeDegree.columns = ['nodeDegree']
joinedDF = nodeDegree.join(avgNeighborDegree)
plt.figure(figsize=[10,8])
plt.plot(joinedDF.nodeDegree, joinedDF.avgNeighborDegree, 'ok', alpha = 0.3)
plt.title('Meta Graph')
plt.ylabel('Average Neighbor Degree')
plt.xlabel('k')
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
```

### 3.4. Node degree vs. clustering coefficient

```python
clusteringCoeff = nx.clustering(G)
clusteringCoeff = pd.DataFrame.from_dict(clusteringCoeff, orient='index')
clusteringCoeff.columns = ['clusteringCoeff']
nodeDegree = nx.degree(G)
nodeDegree = pd.DataFrame.from_dict(nodeDegree, orient='index')
nodeDegree.columns = ['nodeDegree']
joinedDF = nodeDegree.join(clusteringCoeff)
plt.figure(figsize=[10,8])
plt.plot(joinedDF.nodeDegree, joinedDF.clusteringCoeff, 'ok', alpha = 0.6)
plt.title('Meta Graph')
plt.ylabel('Clustering coefficient')
plt.xlabel('k')
plt.grid(True, ls='--', lw=0.5, alpha=0.5, dash_capstyle = 'round', c='gray')
```

### 3.5. Connected components

#### 3.5.1. Counting all connected components

```python
counter = 0

for i in nx.connected_components(G):
    counter+=1
    
print 'Number of connected components constitutes %d' % counter
```

```python
counter = 0

for i in nx.connected_components(G):
    if len(i)>1:
        counter+=1
    
print 'Number of connected components constitutes %d' % counter
```

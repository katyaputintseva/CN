```python
from __future__ import division
import os
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import networkx as nx
import itertools
import math
import seaborn as sns
import distance

%matplotlib inline
```

### 1. Extracting stems from the stem files<br> 
<font color='red'>Extracting only a complementary half of each stem!</font>

```python
aa = 'trp'

df = pd.DataFrame.from_csv('../copies_free_db/' + aa + '/stems.txt', sep = '\t')

all_stems = [(df.Stem_1[y][:7] + df.Stem_2[y][:4] + df.Stem_3[y][:5] + df.Stem_4[y][:5]) for y in range(len(df.Stem_1))]
```

### 2. Creating a dict of tRNA stems and representability in the database
Cleaning the dict from sequences, having anything outside 'ATGC-' (not really sure how to deal with '-'s)

```python
counting_sqs = {}
for sq in all_stems:
    if sq not in counting_sqs.keys():
        counting_sqs[sq]=1
    else:
        counting_sqs[sq]+=1
        
for sq in counting_sqs.keys():
    for nt in sq:
        if nt not in 'AGTC-':
            counting_sqs.pop(sq, None)
            
```

### 3. Creating a pandas table with number of mutations for each pair of tRNAs
Extracting number of pairs on different distances, using value_counts, and writing it into a <i>hamming_dict</i>

```python
hamming_table = pd.DataFrame(columns = counting_sqs.keys(),index = counting_sqs.keys())
counter = 0

for sq1 in counting_sqs.keys():
    for sq2 in counting_sqs.keys():
        counter += 1
        hamming_table.set_value(sq1, sq2, distance.hamming(sq1, sq2))
        if counter % 1000000 == 0:
            print counter
            
hamming_dict = hamming_table.stack().value_counts()

```

### 4. Dividing the number of pairs by two 
(as the numbers are duplicated in the table due to the nature of the table) 

Taking a ln from those numbers.
Writing the results into a <i>hamming_dict_log</i><br>
In case there are no pairs on any distance between 1 and 21, setting the value of that key to 0

```python
hamming_dict_log={}
for key in hamming_dict.keys():
    if key!=0:
        hamming_dict_log[key]=math.log(hamming_dict[key]/len(df))
for i in range(1,22):
    if i not in hamming_dict_log.keys():
        hamming_dict_log[i]=0.0       
        
hamming_dict        
        
```

### 5. Calculating all possible pairs for each distance
Taking a ln and writing it into a <i>hypothetical_dict</i>

```python
hypothetical_dict={}

for n in range(1,22):
    hypothetical_dict[n]=math.log(len(list(itertools.combinations('a'*21,n)))*(3**n))
    
    
```

### 6. Plotting the sequence spaces growth rates

```python
plt.figure(figsize=[25,10])
plt.plot([math.log(i) for i in range(1,22)],[k for k in hypothetical_dict.values()],label='hypothetical',color='#F9C535')
plt.plot([math.log(i) for i in range(1,22)],[k for k in hamming_dict_log.values()],label='visible',color='#3BAEA0')
plt.xlabel('Single source shortest path')
plt.ylabel('Number of such connections, ln')
plt.xticks([math.log(i) for i in range(1,22)])
plt.title(aa)
plt.xlim([0,3.05])
plt.ylim([-5,28])
plt.legend(loc=0);
plt.savefig('../sq_space_growth_rates/' + aa + '.png');

```
